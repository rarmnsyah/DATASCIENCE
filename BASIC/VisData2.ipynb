```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = pd.DataFrame(iris['data'])
target = pd.DataFrame(iris['target'])
```

```python
df.plot.line(title = 'iris datasets')
```

```python
df.plot.hist(subplots = True, layout  = (2,2), bins = 20, figsize = (10,10))
```

```python
df['sepal length (cm)'].groupby(target.squeeze()).mean().plot.bar()
```

```python
import seaborn as sns

sns.lineplot(data = df)
```

```python
sns.kdeplot(data = df)
```
