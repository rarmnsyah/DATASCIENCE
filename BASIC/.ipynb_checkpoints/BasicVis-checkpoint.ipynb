### Using matplotlib pyplot to basic visualized data 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### 1. Plotline

```python
years = [1950, 1960, 1970, 1980, 1990, 2000]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7]

plt.plot(years, gdp, color = 'green', marker = '.', linestyle = 'solid')
```

#### 2. Bar chart

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

plt.bar(movies, num_oscars, color = 'green')
plt.ylabel('num oscars') #set label for y axis
plt.xlabel('movies') #set label for x axis
```

##### 3. Line Charts (More Than one Line) 

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x,y in zip(variance, bias_squared)]
xs = [i + 0.5 for i in range(len(variance))]

plt.plot(xs, variance, 'g-', label = 'variance')
plt.plot(xs, bias_squared, 'b', label = 'bias_squared')
plt.plot(xs, total_error, 'r:', label = 'total error')

plt.legend(loc = 9)
plt.xlabel('model complexity')
```

##### 4. Scatterplot 

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for friend,minute,label in zip(friends, minutes, labels):
    plt.annotate(label,
                xy = (friend,minute), #points annotate
                textcoords = 'offset points', #title 
                xytext = (5,-5) 
                )
    
plt.title('Daily Minutes vs Number of Friends')
plt.xlabel('#friends')
plt.ylabel('Minutes spend')
```
