```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### PROBABILITY DISTRIBUTION

For information, distribution probabilities is generator data, with some paramater.
Example, we have normal distribution with mean = MEAN, and std = STANDART_DEVIATION, when we apply distribution, some data will generated by normal distribution with mean = MEAN, and std = STANDART_DEVIATION.

```python
MEAN = 150
STANDART_DEVIATION = 40
NUM_SAMPLING = 50
NUM_SIZE = 50
```

```python
data = np.random.normal(MEAN, STANDART_DEVIATION, NUM_SIZE)
print(pd.DataFrame(data).describe())
plt.hist(data, label = 'NORMAL DISTRIBUTION')
plt.xlabel('Distributed Value')
plt.ylabel('Frequency Occured')
plt.show()
```

### SAMPLING DISTRIBUTION
Sampling distribution dari sebuah populasi akan menghasilkan data dengan mean = MEAN dan std = STANDART_DEVIATION / sqrt(sample_size). Data yang terdistribusi akan membentuk distribusi normal dengan mean seperti yang telah dijelaskan.

```python
sampling_distribution = []
for i in range(NUM_SAMPLING):
    data = np.random.normal(MEAN, STANDART_DEVIATION, NUM_SIZE)
    sampling_distribution.append(data.mean())
    
sampling_distribution = np.array(sampling_distribution, dtype = 'int')
# sampling_distribution = n(data, sampling_distribution)
    
plt.hist(sampling_distribution, label = 'Sampling Distribution from Normal Distribution Probabilities')
plt.xlabel('mean')
plt.ylabel('frequency mean occured')
plt.show()
    
print(f'Population Mean: {MEAN}, Population STD: {STANDART_DEVIATION}')
print(f'Sampling Mean: {sampling_distribution.mean()}, Sampling STD: {sampling_distribution.std()}')
```

### CONFIDENCE INTERVAL
Confidence interval adalah interval 'confidence' dari sampling distribution berkisar di antara interval tersebut. Contoh confidence interval (95 % / 1std), hal ini berarti mean dari sampling distribution memiliki 'confidence interval' 95% akan berada pada interval mean +- 1SE, dimana SE = std/sqrt(STANDART_DEVIATION).

```python
interval1std = []
interval2std = []
interval3std = []

sampling_std = STANDART_DEVIATION/np.math.sqrt(NUM_SAMPLING)

for i in sampling_distribution:
    if i in range(int(MEAN - 1*sampling_std), int(MEAN + 1*sampling_std)):
        interval1std.append(i)
        interval2std.append(i)
        interval3std.append(i)
    elif i in range(int(MEAN - 2*sampling_std), int(MEAN + 2*sampling_std)):
        interval2std.append(i)
        interval3std.append(i)
    elif i in range(int(MEAN - 3*sampling_std), int(MEAN + 3*sampling_std)):
        interval3std.append(i)
        
print('% interval1std : {}\n% interval2std : {}\n% interval3std : {}'.format(len(interval1std)/len(sampling_distribution), len(interval2std)/len(sampling_distribution),len(interval3std)/len(sampling_distribution)))
    
```

Berdasarkan hal ini kita dapat membuktikan bahwa pada sampling distribution yang telah digenerate. Terdapat sebanyak 64% data berada di interval 1std (normal: 68), 94% berada di interval 2std(normal: 95), dan 100% berada di interval 3std (normal: 99.7). Perbedaan ini diakibatkan n sampling distribution yang tidak terlalu besar (50), namun hal ini dapat memberi kita pengetahuan mendasar mengenai confidence interval.
